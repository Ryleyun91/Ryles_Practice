```python
import json
import copy
```

```python
class InputExample(object):
    """
    A single training/test example for simple sequence classification.

    Args:
        guid: Unique id for the example.
        text_a: string. The untokenized text of the first sequence. For single
        sequence tasks, only this sequence must be specified.
        label: (Optional) string. The label of the example. This should be
        specified for train and dev examples, but not for test examples.
    """

    def __init__(self, guid, text_a, label):
        self.guid = guid
        self.text_a = text_a
        self.label = label

    # 디버깅시 객체를 유의미한 문자열로 출력하는 메소드
    # https://blog.naver.com/websearch/222119748890
    ## __str__(self)는 print(클래스명)을 하면 출력해 줌
    ### 파이썬에서는 str이 없으면 repr를 부르기 때문에 둘 중 하나면 repr를 하는게 좋음
    def __repr__(self):
        return str(self.to_json_string())

    def to_dict(self):
        """Serializes this instance to a Python dictionary."""
        output = copy.deepcopy(self.__dict__)
        return output

    def to_json_string(self):
        """Serializes this instance to a JSON string."""
        return json.dumps(self.to_dict(), indent=2, sort_keys=True) + "\n"
```

```python
example = InputExample('tarin-0', '진정한 쓰레기', 0)
```
