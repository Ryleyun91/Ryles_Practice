```python
def AND(x1, x2):
    if x1 == x2 and x1 == 0:
        return 0
    elif x1 == x2 and x1 == 1:
        return 1
    else:
        return 0  
```

```python
def AND(x1, x2):
    w1, w2, theta= 0.5, 0.5, 0.7
    y = x1*w1 + x2*w2
    
    if y <= theta:
        return 0
    else:
        return 1
```

```python
for xs in [(0, 0), (1, 0), (0, 1), (1, 1)]:
    y = AND(xs[0], xs[1])
    print(str(xs) + " -> " + str(y))
```

# bias

```python
def AND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.7
    y = np.sum(w*x) + b
    
    if y <= 0:
        return 0
    else: 
        return 1
```

```python
for xs in [(0, 0), (1, 0), (0, 1), (1, 1)]:
    y = AND(xs[0], xs[1])
    print(str(xs) + " -> " + str(y))
```

```python
def NAND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([-0.5, -0.5])
    b = 0.7
    y = np.sum(w*x) + b
    
    if y <= 0:
        return 0
    else: 
        return 1
```

```python
for xs in [(0, 0), (1, 0), (0, 1), (1, 1)]:
    y = NAND(xs[0], xs[1])
    print(str(xs) + " -> " + str(y))
```

```python
def OR(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.2
    y = np.sum(w*x) + b
    
    if y <= 0:
        return 0
    else: 
        return 1
```

```python
for xs in [(0, 0), (1, 0), (0, 1), (1, 1)]:
    y = OR(xs[0], xs[1])
    print(str(xs) + " -> " + str(y))
```

# XOR 게이트 구현

```python
def XOR(x1, x2):
    s1 = NAND(x1, x2)
    s2 = OR(x1, x2)
    y = AND(s1, s2)
    
    return y
```

```python
for xs in [(0, 0), (1, 0), (0, 1), (1, 1)]:
    y = XOR(xs[0], xs[1])
    print(str(xs) + " -> " + str(y))
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

# ReLU 함수 구현

```python
def ReLU(x):
    mask = (x <= 0)
    out = x.copy()
    # mask된 요소만 0으로 바꿔주기
    out[mask] = 0
    
    return out
```

```python
def ReLU(x):
    return np.maximum(0, x)
```

```python
x = np.array([2, 1, -1])
ReLU(x)
```

# MLP 구현

```python
import numpy as np

def init_network():
    network = {}
    network['W1'] = np.array([[0.1,0.3,0.5],[0.2,0.4,0.6]])
    network['b1'] = np.array([0.1,0.2,0.3])
    network['W2'] = np.array([[0.1,0.4],[0.2,0.5],[0.3,0.6]])
    network['b2'] = np.array([0.1,0.2])
    network['W3'] = np.array([[0.1,0.3],[0.2,0.4]])
    network['b3'] = np.array([0.1,0.2])
    
    return network

def forward(network, x):
    W1,W2,W3 = network['W1'], network['W2'], network['W3']
    b1,b2,b3 = network['b1'], network['b2'], network['b3']
    
    a1 = np.dot(x, W1) + b1
    z1 = sigmoid(a1)
    a2 = np.dot(z1, W2) + b2
    z2 = sigmoid(a2)
    a3 = np.dot(z2, W3) + b3
    
    y = a3
    
    return y
    
network = init_network()
x = np.array([1.0, 0.5])
y = forward(network, x)
print(y)
```

```python
def load_mnist(normalize=True, flatten=True, one_hot_label=False):
    def _change_one_hot_label(X):
        T = np.zeros((X.size, 10))
        for idx, row in enumerate(T):
            row[X[idx]] = 1

        return T

    with open('mnist.pkl', 'rb') as f:
        dataset = pickle.load(f)
        
    if normalize:
        for key in ('train_img', 'test_img'):
            dataset[key] = dataset[key].astype(np.float32)
            dataset[key] /= 255.0
            
    if one_hot_label:
        dataset['train_label'] = _change_one_hot_label(dataset['train_label'])
        dataset['test_label'] = _change_one_hot_label(dataset['test_label'])    
    
    if not flatten:
         for key in ('train_img', 'test_img'):
            dataset[key] = dataset[key].reshape(-1, 1, 28, 28)

    return (dataset['train_img'], dataset['train_label']), (dataset['test_img'], dataset['test_label']) 
```

```python
import numpy as np
import pickle

(x_train, t_train), (x_test, t_test) = load_mnist(normalize=True, flatten=True, one_hot_label=False)
```

```python
x_train[0].shape
```

```python
def init_network():
    with open("sample_weight.pkl", "rb") as f:
        network = pickle.load(f)
        
    return network
        
def predict(network, x):
    W1,W2,W3 = network['W1'], network['W2'], network['W3']
    b1,b2,b3 = network['b1'], network['b2'], network['b3']
    
    a1 = np.dot(x, W1) + b1
    z1 = sigmoid(a1)
    a2 = np.dot(z1, W2) + b2
    z2 = sigmoid(a2)
    a3 = np.dot(z2, W3) + b3
    
    y = a3
    
    return np.exp(y)/sum(np.exp(y))
    
model = init_network()
```

```python
accuracy_cnt = 0

for i in range(len(x_test)):
    y = predict(model, x_test[i])
    if i==0:
        print(y)
        print(t_test[i])
    p = np.argmax(y)
    
    if p == t_test[i]:
        accuracy_cnt+=1


print("Accuracy:" + str(float(accuracy_cnt) / len(x_test)))
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

def displayData(X,Y):
    fig, ax = plt.subplots(nrows=10, ncols=10, figsize=(15, 15))
    fig.suptitle('Display randomly images of the trainig data set')
    
    for i in range(10):
        for j in range(10):
            ind = np.random.randint(X.shape[0])
            tmp = X[ind, :].reshape(28, 28)
            ax[i, j].set_title("Label: {}".format(Y[ind]))
            ax[i, j].imshow(tmp, cmap='gray_r')
            plt.setp(ax[i, j].get_xticklabels(), visible=False)
            plt.setp(ax[i, j].get_yticklabels(), visible=False)
            
    fig.subplots_adjust(hspace=0.5, wspace=0.5)
            
displayData(x_train,t_train)  
```

# batch

```python
%%time
network = init_network()
accuracy_cnt = 0
batch_size = 100

for i in range(0, len(x_test), batch_size):
    x_batch = x_test[i:i+batch_size]
    t_batch = predict(network, x_batch)
    
    p = np.argmax(t_batch, axis=1)
    
    accuracy_cnt += np.sum(p == t_test[i:i+batch_size])
    

    
print("Accuracy:" + str(float(accuracy_cnt) / len(x_test)))
```
