### 키워드 포함 여부 확인

```python
key_count = []
for keys, text in zip(data["요약문_영문키워드"], data["요약문_기대효과"]):
    count = 0
    for key in str(keys).split(","):
        if str(key).strip() in str(text):
            count+=1
    key_count.append(count)
```

### 띄어쓰기 제거 후 키워드 포함 여부 확인

### 요약문_연구목표/연구내용/기대효과 어절 사전 (형태소 분석 전)

#### 연구목표

```python
import re

goal_words = []

for num in range(0, 46):
    temp = ""
    for text in data["요약문_연구목표"][data["label"]==num]:
        temp+=(" "+str(text))
    count_dic = dict(Counter(str(temp).split()))
    goal_words.append([num, sorted(count_dic.items(), key = lambda x: x[1], reverse=True)])
```

```python
df_goal_words = pd.DataFrame(goal_words)
df_goal_words[:5]
```

#### 연구내용

```python
content_words = []

for num in range(0, 46):
    temp = ""
    for text in data["요약문_연구내용"][data["label"]==num]:
        temp+=(" "+str(text))
    count_dic = dict(Counter(str(temp).split()))
    content_words.append([num, sorted(count_dic.items(), key = lambda x: x[1], reverse=True)])
```

```python
df_content_words = pd.DataFrame(content_words)
df_content_words[:5]
```

#### 기대효과

```python
effect_words = []

for num in range(0, 46):
    temp = ""
    for text in data["요약문_기대효과"][data["label"]==num]:
        temp+=(" "+str(text))
    count_dic = dict(Counter(str(temp).split()))
    effect_words.append([num, sorted(count_dic.items(), key = lambda x: x[1], reverse=True)])
```

```python
df_effect_words = pd.DataFrame(effect_words)
df_effect_words[:5]
```

### 카테고리별 어절 사전 중 키워드 확인하기 

#### 워드 리스트 만들기

```python
from konlpy.tag import Mecab
import re

mecab = Mecab()
goal_word_dict = {i:list() for i in range(46)}
count = 0

for num in range(1, 46):

    for text in data["요약문_기대효과"][data["label"]==num]:
        temp = re.sub("[^ㄱ-힣A-z\s0-9]+|\s+"," ",str(text).strip())
        goal_word_dict[num] += [word for word, pos in mecab.pos(temp) if pos[0] in ["N", "V"]]
    count+=1
    print(count)
```

```python
effect_key = []
for i in goal_word_dict.values():
    effect_key.extend(i)
```

```python
from collections import Counter
goal_key_freq = dict(Counter(effect_key))

with open("effect_key_freq.pkl", 'wb') as f:
    pickle.dump(goal_key_freq, f)
```

```python
goal_word_list = list(goal_word_dict.values())
```
