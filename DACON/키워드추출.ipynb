```python
import pandas as pd
import numpy as np
import re
```

```python
train = pd.read_csv('./open/train.csv')
test = pd.read_csv('./open/test.csv')
```

```python
train_key = list(train['요약문_한글키워드'])
test_key = list(test['요약문_한글키워드'])
```

```python
train_keywords = []
test_keywords = []

for sent in train_key:
    train_keywords.extend(re.split("[,|/;:-_]", re.sub("\s+", "", str(sent))))
    
for sent in test_key:
    test_keywords.extend(re.split("[,|/;:-_]", re.sub("\s+", "", str(sent))))
```

```python
train_keywords = list(train_keywords)
test_keywords = list(test_keywords)
```

```python
from collections import Counter
train_keywords = dict(Counter(train_keywords))
test_keywords = dict(Counter(test_keywords))
```

```python
clean_train = []
for word, freq in train_keywords.items():
    if type(re.search("[\WA-z0-9]+", word))!=re.Match and 20>=len(word)>1 and freq>=20:
        clean_train.append(word)
        
clean_test = []
for word, freq in test_keywords.items():
    if type(re.search("[\WA-z0-9]+", word))!=re.Match and 20>=len(word)>1 and freq>=10:
        clean_test.append(word)
```

```python
clean_keywords = list(set(clean_train + clean_test))
clean_keywords_df = pd.DataFrame(clean_keywords)
```

```python
clean_keywords_df["left-ID"] = None
clean_keywords_df["right-ID"] = None
clean_keywords_df["가중치"] = None
clean_keywords_df["품사"] = "KEY"
clean_keywords_df["의미분류"] = "*"
clean_keywords_df["종성유무"] = ""
clean_keywords_df["읽기"] = clean_keywords_df[0]
clean_keywords_df["타입"] = "*"
clean_keywords_df["첫번째품사"] = "*"
clean_keywords_df["마지막품사"] = "*"
clean_keywords_df["표현"] = "*"
clean_keywords_df["색인표현"] = "*"
```

```python
clean_keywords_df
```

```python
import Korean_jaso as jaso

for idx, word in enumerate(clean_keywords):
    if jaso.jaso(word[-1])[-1]=="없음":
        clean_keywords_df.iloc[idx,6] = "F"
    else:
        clean_keywords_df.iloc[idx,6] = "T"
```
