```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfTransformer,TfidfVectorizer
from collections import defaultdict
from konlpy.tag import Mecab
import pandas as pd
import numpy as np
import re
```

```python
love_poem = pd.read_csv('Poem_Love.csv',lineterminator='\n')
Love_Poem = love_poem['본문'].values.tolist()
clean_lovepoem = []
for j in Love_Poem:
	cleanpoem = re.sub('[…().!?,""---―\ufeff\u200bknsjQuery1319700039250="65">jQuery1319700039250="68">\r\xa0~‘’;/]','',str(j))
	edited_poem = re.sub('[\u2e80-\u2eff\u31c0-\u31ef\u3200-\u32ff\u3400-\u4dbf\u4e00-\u9fbf\uf900-\ufaff]', '', cleanpoem)
	clean_lovepoem.append(edited_poem)
print(clean_lovepoem[:2])
```

```python
data_lovepoem = []
for i in clean_lovepoem:
	splited = i.split('\n')
	data_lovepoem.append(splited)
splited_lovepoem = []
for j in data_lovepoem:
	for k in j:
		if k:
			splited_lovepoem.append(k)
splited_lovepoem[:2]
```

# 핵심키워드 추출(TF-IDF)

```python
mecab = Mecab()
preprocessed_docs = []
for i in splited_lovepoem:
  token_list = [] 
  for token in mecab.pos(i):
    if token[1] in ['NNG']:
      token_list.append(token[0])
  preprocessed_docs.append(" ".join(token_list))
display(preprocessed_docs[:10],len(preprocessed_docs))
```

```python
vectorizer = TfidfVectorizer()
sp_matrix = vectorizer.fit_transform(preprocessed_docs)
word2id = defaultdict(lambda : 0)
for idx, feature in enumerate(vectorizer.get_feature_names()):
    word2id[feature] = idx
keywords_lovelists = []
for i, sent in enumerate(preprocessed_docs):
    print('====== document[%d] ======' % i)
    for token in sent.split():
        matrix = sp_matrix[i, word2id[token]]
        print(token,matrix)
        if 0.3 <= matrix < 1.0:
            keywords_lovelists.append(token)
display(keywords_lovelists[:5],len(keywords_lovelists))
```

```python
key_lovepoem = []
for v in keywords_lovelists:
	if v not in key_lovepoem:
        	key_lovepoem.append(v)
# tests = []
# for i,j in enumerate(key_lovepoem):
#     if i < 5:
#         tests.append(j)
display(key_lovepoem[:5],len(key_lovepoem))
```
