```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfTransformer,TfidfVectorizer
from collections import defaultdict
import pandas as pd
import numpy as np
from konlpy.tag import Mecab
import re
```

```python
poem_corpus = pd.read_csv('korean_nonpup_lines.csv')
corpus = []
for i in poem_corpus['0']:
	cleanPoem = re.sub('[…().!?,---―&&lt;&gt:]','',i)
	real_cleaned = re.sub('[&name1&name2&name3&name4&name5&name6&name7]','',cleanPoem)
	corpus.append(real_cleaned)
corpus
```

```python
mecab = Mecab()
preprocessed_docs = []
for i in corpus:
  token_list = [] 
  for token in mecab.pos(i):
    if token[1] in ['NNG']:
      token_list.append(token[0])
  preprocessed_docs.append(" ".join(token_list))
display(preprocessed_docs[:10],len(preprocessed_docs))
```

```python
vectorizer = TfidfVectorizer()
sp_matrix = vectorizer.fit_transform(preprocessed_docs)
word2id = defaultdict(lambda : 0)
for idx, feature in enumerate(vectorizer.get_feature_names()):
    word2id[feature] = idx
keywords_lists = []
for i, sent in enumerate(preprocessed_docs):
    print('====== document[%d] ======' % i)
    for token in sent.split():
        matrix = sp_matrix[i, word2id[token]]
        print(token,matrix)
        if 0.3 <= matrix < 1.0:
            keywords_lists.append(token)
print(keywords_lists[:10])
```

```python
key_lovepoem = [] 
for v in keywords_lists:
	if v not in key_lovepoem:
        	key_lovepoem.append(v)
print(key_lovepoem)
```
